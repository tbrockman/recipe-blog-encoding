Hide secret messages in natural-looking recipe blog introductions using neural linguistic steganography.

This notebook lets you try encoding and decoding secrets using `Qwen/Qwen2.5-3B` on a free Colab GPU — no model approvals or API keys required.

## Encode

Hide a secret message inside a recipe blog introduction. Change `MESSAGE` to whatever you'd like to encode.

```python
MODEL = "Qwen/Qwen2.5-3B"
MESSAGE = "attack at dawn"

!python main.py encode \
  --model {MODEL} \
  --message "{MESSAGE}" \
  --stego-file stego_output.txt
```

## Decode

Recover the hidden message from the generated text. Uses the same model and prompt as encoding.

```python
!python main.py decode \
  --model {MODEL} \
  --stego-file stego_output.txt
```

## Custom prompt

You can use any prompt — the recipe framing is just the default. Both sides need the exact same prompt to encode/decode.

```python
CUSTOM_PROMPT = "The best thing about making scrambled eggs is: "
CUSTOM_MESSAGE = "meet me at noon"

!python main.py encode \
  --model {MODEL} \
  --message "{CUSTOM_MESSAGE}" \
  --prompt "{CUSTOM_PROMPT}" \
  --stego-file custom_output.txt
```
